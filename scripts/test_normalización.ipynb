```python
import qlib
from qlib.data import D

# 1. Inicializa con la ruta absoluta para evitar errores de path
qlib.init(provider_uri="/home/toni/.qlib/qlib_data/us_data")

# 2. Carga datos usando argumentos con nombre (evita el error de inst_processors)
df = D.features(
    instruments=['AAPL'], 
    fields=['$close', '$open', '$high', '$low'], 
    start_time='2026-03-25', 
    end_time='2026-03-31', 
    freq='day' # Asegúrate de especificar la frecuencia aquí
)

print(df)
```

```python
df = D.features(["AAPL"], ["$close", "$factor"], start_time="2026-03-30", end_time="2026-03-31")
df["raw_close_reconstruido"] = df["$close"] / df["$factor"]
print(df)
```
